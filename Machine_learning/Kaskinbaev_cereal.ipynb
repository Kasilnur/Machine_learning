### Импортирование библиотек

```python
import pandas as pd   
import seaborn as sns    
import matplotlib.pyplot as plt
%matplotlib inline
```

### Чтение и просмотр данных

```python
# Присваиваем имя датасету
cereal = pd.read_csv('cereal.csv')
cereal
```

В наборе данных 16 столбцов и 77 строк. Типы данных в них - `object`, `int`, `float`.
В этой таблице записаны различные типы хлопьев, имя которых записываются в столбец `name`.
* `mfr` - производитель
* `type` - тип зерна
* `calories` - кол-во калорий
* `protein` - кол-во белков
* `fat` - кол-во жиров
* `sodium` - кол-во натрия
* `fiber` - кол-во клетчатки
* `carbo` - кол-во карбо
* `sugars` - кол-во сахара
* `potass` - кол-во калия
* `vitamins` - кол-во витаминов
* `shelf` - полка хранения
* `weight` - вес
* `cups` - порция
* `rating` - рейтинг зерна

```python
cereal.info()
```

Набор данных не очень большой: 77 записей и 16 признаков. Отсутсвуют пустые значения.

```python
cereal.describe()
```

Аномальных значений нет, за исключением столбцов, `carbo`, `sugars` и `potass`: в них присутсвуют отрицательные значения.

```python
cereal.duplicated().sum()
```

Отсутвуют дубликаты

### Предобработка данных

```python
# Заменим отрицательные числа медианными значениями
med_carbo = cereal[cereal['carbo'] > 0]['carbo'].median()
med_sugars = cereal[cereal['sugars'] > 0]['sugars'].median()
med_potass = cereal[cereal['potass'] > 0]['potass'].median()
cereal[cereal[['potass']] < 0] = med_potass
cereal[cereal[['sugars']] < 0] = med_sugars
cereal[cereal[['carbo']] < 0] = med_carbo
# Проверяем изменения
cereal.describe()
```

Данные обработаны. Аномальных значений нет

### Анализ данных

```python
cereal.hist(figsize=(10,10), bins=40);
```

```python
# Популярные хлопья по размеру порции и полке.
sns.pairplot(cereal[['shelf', 'cups','rating']]);
```

```python
# Коррелированный датасет
df = cereal[['calories','protein','fat','sodium','fiber','carbo','sugars','potass','vitamins','shelf','cups','rating']].corr()
df
```

```python
# От чего зависит рейтинг хлопьев
plt.figure(figsize=(14,6))
sns.heatmap(df, annot=True);
```

```python
# Рейтинг производителей
sns.barplot(data=cereal, x=cereal['rating'], y=cereal['mfr']);
```

```python
# Какой тип хлопьев пользуется популярностью
sns.displot(cereal['type']);
```
